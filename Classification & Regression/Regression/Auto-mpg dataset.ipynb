```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
from math import sqrt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression

```

Load the data set

```python
data = pd.read_csv('C:\\Users\\Yuvraj\\Desktop\\Data Science\\kaggle\\Data\\auto-mpg.csv' )
data.head()
```

Replace missing values with NaN

```python
data = data.replace('?', np.nan)
```

Drop rows with missing values

```python

data = data.dropna()
```

Print the summary of the DataFrame

```python
data.info
```

Print the statistical summary of the DataFrame

```python
data.describe() 
```

Count the number of missing values in each column

```python
data.isnull().sum()
```

Count the number of occurrences of each unique value

```python
data["car name"].value_counts()
```

replace missing values in the "horsepower" column

```python
data["horsepower"]=data["horsepower"].replace("?",1)
data["horsepower"]=data["horsepower"].astype(int)
```

LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
data["car name"]=l.fit_transform(data["car name"])
```

DataFrame used to extract rows and columns 

```python
X = data[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin']]
y = data['mpg']
```

Print X data

```python
X
```

Print y data

```python
y
```

```python
data
```

Split the data into training and testing sets

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2,random_state=11)
X_train
```

Train the Linear Regression model

```python
reg = LinearRegression().fit(X_train, y_train)
```

Make predictions on the test set 

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

Calculate RMSE

```python

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

```python
from sklearn.metrics import mean_squared_error, r2_score


mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("R-squared:", r2)
```
